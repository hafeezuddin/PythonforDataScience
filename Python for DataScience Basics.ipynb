# GUESS THE WORD GAME WITH WITHIN 4 ATTEMPTS

```python
a = input("Enter your Guess:")
counter = 1
Correct_Answer = "tiger"

if a == Correct_Answer:
    print("Correct Answer")
else:
    while a!="tiger" and counter<3:
        print("Your Guess is wrong, Re-enter it again",counter)
        
        a = input("Enter your Guess:")
        counter = counter+1
        if a == Correct_Answer:
            print("Correct Answer")
            break
        if a!=Correct_Answer and counter==3:
            print("Exhausted")
            break
        else:
            print("Wrong Answer")
```

```python
def universal(*args):            #takes n number of arguments  *args acts as a list
    sum = 0
    for i in range(len(args)):   #iterating through each element
        sum = sum+args[i]
    return sum
universal(2,3,4)
```
